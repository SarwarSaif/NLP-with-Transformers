## Alignment with Dot-product


image.png

When performing NLP tasks, typically can be converted into Word2Vec and in the Transformers a similar operation is perfomed to get a Vector that allows us to represent meaning Numerically.
When working these Vectors, we would expect sentences with similar meaning to have a similar set of values. 
When we iterate through each word and compare individual vectors between the two sequences/sentences, we should find that similar words like "Hello" and "Ciao" will have Higher Attention Activation, whereas dissimilar words like "Hello" and "Come" have a Lower Activation.
This relationship can be visualize a little bit better using a heatmap.


image.png

In the heatmap, the darker the block the higher the similarity. The similarity between these word vectors is known as Alignment. This is different to Attention, so we first calculate alignment between two vectors using the DOT product. 

```python
hello = [0.71, 0.14, 0.51]
hi = [0.69, 0.15, 0.48]
tomato = [0.16, 0.59, 0.49]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16, 12))
ax = fig.add_subplot(projection='3d')

xs = [hello[0], hi[0], tomato[0]]
ys = [hello[1], hi[1], tomato[1]]
zs = [hello[2], hi[2], tomato[2]]

ax.scatter(xs, ys, zs, s=100)

ax.set_xlim((0, 1))
ax.set_ylim((0, 1))
ax.set_zlim((0, 1))
```

```python
import numpy as np
a = np.array(hello)
b = np.array(hi)
c = np.array(tomato)
```

```python
np.matmul(a, b.T), np.dot(a, b)
```

```python
np.matmul(a, b.T), np.matmul(b, c.T), np.matmul(c, a.T) 
# a and b has the maximum score due to their similarity
```

```python
np.matmul(a, a.T) 
# Even though the score with the perfect allignment should have been "1"
# But dot product doesn't give exact 1. In fact for larger values dot product is not that great
# In that Cosine gives much better result.
```
